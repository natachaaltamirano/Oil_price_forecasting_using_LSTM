```python
n_t=800
```

```python
def timeseries_to_supervised(data, lag=1):
    df = DataFrame(data)
    columns = [df.shift(i) for i in range(1, lag+1)]
    columns.append(df)
    df = concat(columns, axis=1)
    return df
 
```

```python
features=1
supervised = timeseries_to_supervised(series.values, features)
supervised_values = supervised.values[features:,:]
```

```python
train, test = supervised_values[0:-n_t, :], supervised_values[-n_t:, :]

```

```python
def scale(train, test):
    # fit scaler
    scaler = MinMaxScaler(feature_range=(-1, 1))
    scaler = scaler.fit(train)
    # transform train
    train = train.reshape(train.shape[0], train.shape[1])
    train_scaled = scaler.transform(train)
    # transform test
    test = test.reshape(test.shape[0], test.shape[1])
    test_scaled = scaler.transform(test)
    return scaler, train_scaled, test_scaled
 
```

```python
scaler, train_scaled, test_scaled = scale(train, test)
```

## Changing number of neurons

```python


for num_neurons in [1,5,10,20,50]:
    
    X, y = train_scaled[:, 0:-1], train_scaled[:, -1]
    test_X,test_y=test_scaled[:, 0:-1], test_scaled[:, -1]
    X = X.reshape(X.shape[0], 1, X.shape[1])
    test_X=test_X.reshape(test_X.shape[0], 1, test_X.shape[1])
    
    
    model = Sequential()
    model.add(LSTM(num_neurons, input_shape=(X.shape[1], X.shape[2]),
                  bias_regularizer= L1L2(l1=0.00, l2=0.00) ,
                       kernel_regularizer=L1L2(l1=0.00, l2=0.00)))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')

    # Fitting the model
    history = model.fit(X, y,
                        epochs=20,
                        batch_size=10, 
                        validation_data=(test_X, test_y), 
                        verbose=0, 
                        shuffle=False)
    # Plotting the training progression
    pyplot.plot(history.history['loss'], label='train')
    pyplot.plot(history.history['val_loss'], label='test')
    pyplot.legend()
    pyplot.title("Num of Neurons:" + str(num_neurons))
    pyplot.xlabel("Epoch")
    pyplot.ylabel("Loss")
    pyplot.show()
    
    yhat = model.predict(test_X)
    test_X = test_X.reshape((test_X.shape[0], test_X.shape[2]))
    inv_yhat = concatenate((test_X[:, 0:], yhat), axis=1)
    inv_yhat = scaler.inverse_transform(inv_yhat) 
    inv_yhat = inv_yhat[:, features*1]
    test_y = test_y.reshape((len(test_y), 1))
    inv_y = concatenate((test_X[:, 0:], test_y), axis=1) 
    inv_y = scaler.inverse_transform(inv_y)
    inv_y = inv_y[:, features*1]
    rmse = np.sqrt(mean_squared_error(inv_y, inv_yhat)) 
    
    pyplot.plot(inv_y, label = 'Actual')
    pyplot.plot(inv_yhat, label = 'Predicted')
    pyplot.title("Num of Neurons:" + str(num_neurons))
    pyplot.legend()
    pyplot.show()
    
    
    print("----------------------------------------------------------------")
```

### Discoveries

- The more number of neurons the more it overfitts, and the less number of epochs that are needed to reach this stage

## Changing the batch size with fized num_neurons=5

```python


for batchs in [5,10,20,50]:
    
    X, y = train_scaled[:, 0:-1], train_scaled[:, -1]
    test_X,test_y=test_scaled[:, 0:-1], test_scaled[:, -1]
    X = X.reshape(X.shape[0], 1, X.shape[1])
    test_X=test_X.reshape(test_X.shape[0], 1, test_X.shape[1])
    
    
    model = Sequential()
    model.add(LSTM(5, input_shape=(X.shape[1], X.shape[2]),
                  bias_regularizer= L1L2(l1=0.00, l2=0.00) ,
                       kernel_regularizer=L1L2(l1=0.00, l2=0.00)))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')

    # Fitting the model
    history = model.fit(X, y,
                        epochs=10,
                        batch_size=batchs, 
                        validation_data=(test_X, test_y), 
                        verbose=0, 
                        shuffle=False)
    # Plotting the training progression
    pyplot.plot(history.history['loss'], label='train')
    pyplot.plot(history.history['val_loss'], label='test')
    pyplot.legend()
    pyplot.title("Batch size:" + str(batchs))
    pyplot.xlabel("Epoch")
    pyplot.ylabel("Loss")
    pyplot.show()
    
    yhat = model.predict(test_X)
    test_X = test_X.reshape((test_X.shape[0], test_X.shape[2]))
    inv_yhat = concatenate((test_X[:, 0:], yhat), axis=1)
    inv_yhat = scaler.inverse_transform(inv_yhat) 
    inv_yhat = inv_yhat[:, features*1]
    test_y = test_y.reshape((len(test_y), 1))
    inv_y = concatenate((test_X[:, 0:], test_y), axis=1) 
    inv_y = scaler.inverse_transform(inv_y)
    inv_y = inv_y[:, features*1]
    rmse = np.sqrt(mean_squared_error(inv_y, inv_yhat)) 
    
    pyplot.plot(inv_y, label = 'Actual')
    pyplot.plot(inv_yhat, label = 'Predicted')
    pyplot.title("Batch size:" + str(batchs))
    pyplot.legend()
    pyplot.show()
    
    
    print("----------------------------------------------------------------")
```

### Discoveries

- The larger the batch size the more it overfitts and the longer the epochs it needs to reach this stage

## Changing the regularization with num_neuros=5, batch_size=10

```python


#Bias regularization

X, y = train_scaled[:, 0:-1], train_scaled[:, -1]
test_X,test_y=test_scaled[:, 0:-1], test_scaled[:, -1]
X = X.reshape(X.shape[0], 1, X.shape[1])
test_X=test_X.reshape(test_X.shape[0], 1, test_X.shape[1])


model = Sequential()
model.add(LSTM(5, input_shape=(X.shape[1], X.shape[2]),
                  bias_regularizer= L1L2(l1=0.01, l2=0.01) ,
                       kernel_regularizer=L1L2(l1=0.0, l2=0.0)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# Fitting the model
history = model.fit(X, y,
                    epochs=10,
                    batch_size=10, 
                    validation_data=(test_X, test_y), 
                    verbose=0, 
                    shuffle=False)
# Plotting the training progression
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.title("Bias regularizer:")
pyplot.xlabel("Epoch")
pyplot.ylabel("Loss")
pyplot.show()

yhat = model.predict(test_X)
test_X = test_X.reshape((test_X.shape[0], test_X.shape[2]))
inv_yhat = concatenate((test_X[:, 0:], yhat), axis=1)
inv_yhat = scaler.inverse_transform(inv_yhat) 
inv_yhat = inv_yhat[:, features*1]
test_y = test_y.reshape((len(test_y), 1))
inv_y = concatenate((test_X[:, 0:], test_y), axis=1) 
inv_y = scaler.inverse_transform(inv_y)
inv_y = inv_y[:, features*1]
rmse = np.sqrt(mean_squared_error(inv_y, inv_yhat)) 

pyplot.plot(inv_y, label = 'Actual')
pyplot.plot(inv_yhat, label = 'Predicted')
pyplot.title("Bias regularizer:")
pyplot.legend()
pyplot.show()

```

```python


#Kernel Regularization


X, y = train_scaled[:, 0:-1], train_scaled[:, -1]
test_X,test_y=test_scaled[:, 0:-1], test_scaled[:, -1]
X = X.reshape(X.shape[0], 1, X.shape[1])
test_X=test_X.reshape(test_X.shape[0], 1, test_X.shape[1])


model = Sequential()
model.add(LSTM(5, input_shape=(X.shape[1], X.shape[2]),
                  bias_regularizer= L1L2(l1=0.0, l2=0.0) ,
                       kernel_regularizer=L1L2(l1=0.01, l2=0.01)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# Fitting the model
history = model.fit(X, y,
                    epochs=10,
                    batch_size=10, 
                    validation_data=(test_X, test_y), 
                    verbose=0, 
                    shuffle=False)
# Plotting the training progression
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.title("Kernel regularizer:")
pyplot.xlabel("Epoch")
pyplot.ylabel("Loss")
pyplot.show()

yhat = model.predict(test_X)
test_X = test_X.reshape((test_X.shape[0], test_X.shape[2]))
inv_yhat = concatenate((test_X[:, 0:], yhat), axis=1)
inv_yhat = scaler.inverse_transform(inv_yhat) 
inv_yhat = inv_yhat[:, features*1]
test_y = test_y.reshape((len(test_y), 1))
inv_y = concatenate((test_X[:, 0:], test_y), axis=1) 
inv_y = scaler.inverse_transform(inv_y)
inv_y = inv_y[:, features*1]
rmse = np.sqrt(mean_squared_error(inv_y, inv_yhat)) 

pyplot.plot(inv_y, label = 'Actual')
pyplot.plot(inv_yhat, label = 'Predicted')
pyplot.title("Kernel regularizer:")
pyplot.legend()
pyplot.show()


```

```python

#Kernel & Bias Regularization


X, y = train_scaled[:, 0:-1], train_scaled[:, -1]
test_X,test_y=test_scaled[:, 0:-1], test_scaled[:, -1]
X = X.reshape(X.shape[0], 1, X.shape[1])
test_X=test_X.reshape(test_X.shape[0], 1, test_X.shape[1])


model = Sequential()
model.add(LSTM(5, input_shape=(X.shape[1], X.shape[2]),
                  bias_regularizer= L1L2(l1=0.1, l2=0.1) ,
                       kernel_regularizer=L1L2(l1=0.01, l2=0.01)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# Fitting the model
history = model.fit(X, y,
                    epochs=10,
                    batch_size=10, 
                    validation_data=(test_X, test_y), 
                    verbose=0, 
                    shuffle=False)
# Plotting the training progression
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.title("Kernel regularizer:")
pyplot.xlabel("Epoch")
pyplot.ylabel("Loss")
pyplot.show()

yhat = model.predict(test_X)
test_X = test_X.reshape((test_X.shape[0], test_X.shape[2]))
inv_yhat = concatenate((test_X[:, 0:], yhat), axis=1)
inv_yhat = scaler.inverse_transform(inv_yhat) 
inv_yhat = inv_yhat[:, features*1]
test_y = test_y.reshape((len(test_y), 1))
inv_y = concatenate((test_X[:, 0:], test_y), axis=1) 
inv_y = scaler.inverse_transform(inv_y)
inv_y = inv_y[:, features*1]
rmse = np.sqrt(mean_squared_error(inv_y, inv_yhat)) 

pyplot.plot(inv_y, label = 'Actual')
pyplot.plot(inv_yhat, label = 'Predicted')
pyplot.title("Kernel regularizer:")
pyplot.legend()
pyplot.show()

```

### Discoveries

- Bias regularization in general shifts the model up
- Kernel regularization changes the extreme of the tails up/down
- The overall effect is to extremelly shift up

```python
def prediction(a):
    return model.predict(np.array([[[a]]]))[0,0]
```

```python
def rec_pre(a,n):
    if n==0:
        return prediction(a)
    else:
        return rec_pre(prediction(a),n-1)
 
```
