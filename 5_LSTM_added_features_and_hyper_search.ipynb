```python
% matplotlib inline

from numpy import concatenate
from pandas import read_csv
from pandas import DataFrame
from pandas import concat
from matplotlib import pyplot
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
import numpy as np


from keras.regularizers import L1L2
```

```python
n_t=800
```

```python
def timeseries_to_supervised(data, lag=1):
    df = DataFrame(data)
    columns = [df.shift(i) for i in range(1, lag+1)]
    columns.append(df)
    df = concat(columns, axis=1)
    return df
 
```

```python
def scale(train, test):
    # fit scaler
    scaler = MinMaxScaler(feature_range=(-1, 1))
    scaler = scaler.fit(train)
    # transform train
    train = train.reshape(train.shape[0], train.shape[1])
    train_scaled = scaler.transform(train)
    # transform test
    test = test.reshape(test.shape[0], test.shape[1])
    test_scaled = scaler.transform(test)
    return scaler, train_scaled, test_scaled
 
```

## Changing the number of features

```python

# Test number of features

for num_features in [1,5,10,20]:
    
    features=num_features
    supervised = timeseries_to_supervised(series.values, features)
    supervised_values = supervised.values[features:,:]
    
    train, test = supervised_values[0:-n_t, :], supervised_values[-n_t:, :]
    
    scaler, train_scaled, test_scaled = scale(train, test)
    
    X, y = train_scaled[:, 0:-1], train_scaled[:, -1]
    test_X,test_y=test_scaled[:, 0:-1], test_scaled[:, -1]
    X = X.reshape(X.shape[0], 1, X.shape[1])
    test_X=test_X.reshape(test_X.shape[0], 1, test_X.shape[1])
    
    
    model = Sequential()
    model.add(LSTM(5, input_shape=(X.shape[1], X.shape[2]),
                  bias_regularizer= L1L2(l1=0.01, l2=0.01) ,
                       kernel_regularizer=L1L2(l1=0.01, l2=0.01)))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')

    # Fitting the model
    history = model.fit(X, y,
                        epochs=20,
                        batch_size=10, 
                        validation_data=(test_X, test_y), 
                        verbose=0, 
                        shuffle=False)
    # Plotting the training progression
    pyplot.plot(history.history['loss'], label='train')
    pyplot.plot(history.history['val_loss'], label='test')
    pyplot.legend()
    pyplot.title("Num of Features:" + str(num_features))
    pyplot.xlabel("Epoch")
    pyplot.ylabel("Loss")
    pyplot.show()
    
    yhat = model.predict(test_X)
    test_X = test_X.reshape((test_X.shape[0], test_X.shape[2]))
    inv_yhat = concatenate((test_X[:, 0:], yhat), axis=1)
    inv_yhat = scaler.inverse_transform(inv_yhat) 
    inv_yhat = inv_yhat[:, features*1]
    test_y = test_y.reshape((len(test_y), 1))
    inv_y = concatenate((test_X[:, 0:], test_y), axis=1) 
    inv_y = scaler.inverse_transform(inv_y)
    inv_y = inv_y[:, features*1]
    rmse = np.sqrt(mean_squared_error(inv_y, inv_yhat)) 
    
    pyplot.plot(inv_y, label = 'Actual')
    pyplot.plot(inv_yhat, label = 'Predicted')
    pyplot.title("Num of features:" + str(num_features))
    pyplot.legend()
    pyplot.show()
```

### Discoveries

- As thought the number of features afect the short scales (large) frequency modes
- I like model that doesn't overfit the short scales
- 20 features seams to be a good number 

## Changing the number of neurons

```python

# Test number of neurons

# Only plotting the last 100 predictions to asses better the overfitting/underfitting

for num_neurons in [1,3,5,10]:
    
    features=20
    supervised = timeseries_to_supervised(series.values, features)
    supervised_values = supervised.values[features:,:]
    
    train, test = supervised_values[0:-n_t, :], supervised_values[-n_t:, :]
    
    scaler, train_scaled, test_scaled = scale(train, test)
    
    X, y = train_scaled[:, 0:-1], train_scaled[:, -1]
    test_X,test_y=test_scaled[:, 0:-1], test_scaled[:, -1]
    X = X.reshape(X.shape[0], 1, X.shape[1])
    test_X=test_X.reshape(test_X.shape[0], 1, test_X.shape[1])
    
    
    model = Sequential()
    model.add(LSTM(num_neurons, input_shape=(X.shape[1], X.shape[2]),
                  bias_regularizer= L1L2(l1=0.01, l2=0.01) ,
                       kernel_regularizer=L1L2(l1=0.01, l2=0.01)))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')

    # Fitting the model
    history = model.fit(X, y,
                        epochs=20,
                        batch_size=10, 
                        validation_data=(test_X, test_y), 
                        verbose=0, 
                        shuffle=False)
    # Plotting the training progression
    pyplot.plot(history.history['loss'], label='train')
    pyplot.plot(history.history['val_loss'], label='test')
    pyplot.legend()
    pyplot.title("Num of neurons:" + str(num_neurons))
    pyplot.xlabel("Epoch")
    pyplot.ylabel("Loss")
    pyplot.show()
    
    yhat = model.predict(test_X)
    test_X = test_X.reshape((test_X.shape[0], test_X.shape[2]))
    inv_yhat = concatenate((test_X[:, 0:], yhat), axis=1)
    inv_yhat = scaler.inverse_transform(inv_yhat) 
    inv_yhat = inv_yhat[:, features*1]
    test_y = test_y.reshape((len(test_y), 1))
    inv_y = concatenate((test_X[:, 0:], test_y), axis=1) 
    inv_y = scaler.inverse_transform(inv_y)
    inv_y = inv_y[:, features*1]
    rmse = np.sqrt(mean_squared_error(inv_y, inv_yhat)) 
    
    pyplot.plot(inv_y[-100:], label = 'Actual')
    pyplot.plot(inv_yhat[-100:], label = 'Predicted')
    pyplot.title("Num of neurons:" + str(num_neurons))
    pyplot.legend()
    pyplot.show()
    
    print("----------------------------------------------------------------")
```

### Discoveries

- The number of neurons give more learning power to the LSTM and thus starts to overfitt and regain some structure

- A good number is possibly between 10

```python
features=20
supervised = timeseries_to_supervised(series.values, features)
supervised_values = supervised.values[features:,:]

train, test = supervised_values[0:-n_t, :], supervised_values[-n_t:, :]

scaler, train_scaled, test_scaled = scale(train, test)

X, y = train_scaled[:, 0:-1], train_scaled[:, -1]
test_X,test_y=test_scaled[:, 0:-1], test_scaled[:, -1]
X = X.reshape(X.shape[0], 1, X.shape[1])
test_X=test_X.reshape(test_X.shape[0], 1, test_X.shape[1])

num_neurons=10
model = Sequential()
model.add(LSTM(num_neurons, input_shape=(X.shape[1], X.shape[2]),
              bias_regularizer= L1L2(l1=0.01, l2=0.01) ,
                   kernel_regularizer=L1L2(l1=0.01, l2=0.01)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')

# Fitting the model
history = model.fit(X, y,
                    epochs=20,
                    batch_size=10, 
                    validation_data=(test_X, test_y), 
                    verbose=0, 
                    shuffle=False)
# Plotting the training progression
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.title("Num of neurons:" + str(num_neurons))
pyplot.xlabel("Epoch")
pyplot.ylabel("Loss")
pyplot.show()

yhat = model.predict(test_X)
test_X = test_X.reshape((test_X.shape[0], test_X.shape[2]))
inv_yhat = concatenate((test_X[:, 0:], yhat), axis=1)
inv_yhat = scaler.inverse_transform(inv_yhat) 
inv_yhat = inv_yhat[:, features*1]
test_y = test_y.reshape((len(test_y), 1))
inv_y = concatenate((test_X[:, 0:], test_y), axis=1) 
inv_y = scaler.inverse_transform(inv_y)
inv_y = inv_y[:, features*1]
rmse = np.sqrt(mean_squared_error(inv_y, inv_yhat)) 

pyplot.plot(inv_y, label = 'Actual')
pyplot.plot(inv_yhat, label = 'Predicted')
pyplot.title("Num of neurons:" + str(num_neurons))
pyplot.legend()
pyplot.show()
```

```python
def prediction_array(a):
    aa=model.predict(np.array([a]))
    
    #eliminates first element
    #a[:,1:]
    #appends a number at the end and protects the format
    b=np.array([np.append(a[:,1:],aa)])
    
    return b
```

```python
def rec_pre(a,n):
    if n==0:
        return prediction_array(a)
    else:
        return rec_pre(prediction_array(a),n-1)
 
```
