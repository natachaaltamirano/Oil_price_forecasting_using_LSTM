```python
from pandas import TimeGrouper


groups = series.groupby(TimeGrouper('A'))
years = {}
for name, group in groups:
    years[name.year] = group.values
    print(len(group.values))
#years.boxplot()
#pyplot.show()
```

```python
from pandas import concat
#series = Series.from_csv('daily-minimum-temperatures.csv', header=0)
i=1

fig = plt.figure(figsize=(15,10))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for year in range(2004,2018):
    #f=plt.figure(figsize=(10,10))
    #plt.subplot(5, 3, i)
    ax = fig.add_subplot(3, 5, i)
    #print(i)
    one_year = series[str(year)]
    groups = one_year.groupby(TimeGrouper('M'))
    months = concat([pd.DataFrame(x[1].values) for x in groups], axis=1)
    months = pd.DataFrame(months)
    months.columns = range(1,13)
    ax= months.boxplot()
    ax.set_title("Year"+ str(year))    
    pyplot.xlabel("month")
    pyplot.ylabel("price")
    i+=1
plt.savefig("seasonality.pdf")    
pyplot.show()
#plt.savefig("seasonality.pdf")
```

```python
from pandas.tools.plotting import lag_plot
#series = Series.from_csv('daily-minimum-temperatures.csv', header=0)
lag_plot(series)
```

```python
from pandas.plotting import scatter_matrix
#series = Series.from_csv('daily-minimum-temperatures.csv', header=0)
values =pd.DataFrame(series.values)
lags = 8
columns = [values]
for i in range(1,(lags + 1)):
    columns.append(values.shift(i))
dataframe = concat(columns, axis=1)
columns = ['t+1']
for i in range(1,(lags + 1)):
    columns.append('t-' + str(i))
dataframe.columns = columns
pyplot.figure(1,figsize=(15,8))
for i in range(1,(lags + 1)):
    ax = pyplot.subplot(240 + i)
    ax.set_title('t+1 vs t-' + str(i))
    #ax.set_xlabel("price")
    #ax.set_ylabel("price")
    
    pyplot.scatter(x=dataframe['t+1'].values, y=dataframe['t-'+str(i)].values)
    
pyplot.savefig("auto.pdf")
pyplot.show()
```

```python
from pandas.tools.plotting import autocorrelation_plot

#series = Series.from_csv('daily-minimum-temperatures.csv', header=0)
autocorrelation_plot(series)
plt.savefig("autocor.pdf")
pyplot.show()
```
